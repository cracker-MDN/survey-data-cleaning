```python
import pandas as pd
import numpy as np

df = pd.read_csv('survey_data_messy.csv')

print("Shape:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
df.head(10)
```

```python
print(df['gender'].unique())
```

```python
# strip whitespace
df['gender'] = df['gender'].str.strip()
```

```python
print(df['gender'].unique())
```

```python
# mapping dictionary
gender_mapping = {
    'MALE': 'Male',
    'M': 'Male',
    'male': 'Male',
    'm': 'Male',
    
    'female': 'Female',
    'f': 'Female',
    'F': 'Female',
    'FEMALE': 'Female'
}
```

```python
df['gender'] = df['gender'].replace(gender_mapping)
print(df['gender'].unique())
```

```python
# employment status column 
print(df['employment_status'].unique())
```

```python
# strip whitespace 
df['employment_status'] = df['employment_status'].str.strip()
```

```python
df['employment_status'].unique()
```

```python
# mapping dictionary
employ_status_mapping = {
    'full-time': 'Full-time',
    'Full Time': 'Full-time',
    'Fulltime': 'Full-time',
    'full time': 'Full-time',
    
    'Part time': 'Part-time',
    
    'Contractor': 'Contract',
    'Freelance': 'Contract',
    
    'intern': 'Intern'
}
```

```python
df['employment_status'] = df['employment_status'].replace(employ_status_mapping)
print(df['employment_status'].unique())
```

```python
# department column 
print(df['department'].unique())
```

```python
# strip whitespace
df['department'] = df['department'].str.strip() 
```

```python
# mapping dictionary
department_mapping = {
    'sales': 'Sales',
    'Sales Team': 'Sales',
    'sales department': 'Sales',
    
    'marketing': 'Marketing',
    'Mktg': 'Marketing',
    'Mkt': 'Marketing',
    'mktg': 'Marketing',
    'Mrkting': 'Marketing',
    'Marketng': 'Marketing',
    'marketing department': 'Marketing',
    
    'I.T.': 'IT',
    'Tech': 'IT',
    'Information Technology': 'IT',
    'IT Department': 'IT',
    'information technology': 'IT',
    
    'HR': 'HR',
    'hr': 'HR',
    'Human Resources': 'HR',
    'Human Resource': 'HR',
    'HR Dept': 'HR',
    
    'R and D': 'R&D',
    'R & D': 'R&D',
    'Research & Development': 'R&D',
    'Research and Development': 'R&D',
    'Research': 'R&D'
}
```

```python
df['department'] = df['department'].replace(department_mapping)
print(df['department'].unique())
```

```python
# satisfaction column
print(df['satisfaction'].unique())
```

```python
df['satisfaction'] = df['satisfaction'].str.strip() 
```

```python
# mapping dictionary
satisfaction_mapping = {
    'very satisfied': 'Very Satisfied', 

    'satisfied': 'Satisfied', 
    'SATISFIED': 'Satisfied',
}
```

```python
df['satisfaction'] = df['satisfaction'].replace(satisfaction_mapping)
print(df['satisfaction'].unique())
```

```python
# recommend column 
print(df['recommend'].unique())
```

```python
print(df['recommend'].isna().sum())
```

```python
# strip whitespace
df['recommend'] = df['recommend'].str.strip()

```

```python
recommend_mapping = {
    'yes': 'Yes', 
    'YES': 'Yes', 
    'Y': 'Yes',

    'maybe': 'Maybe', 

    'N': 'No',
    'no': 'No', 
}
```

```python
df['recommend'] = df['recommend'].replace(recommend_mapping)
print(df['recommend'].unique()) 
```

```python
print(df['respondent_name'].head(20))
```

```python
# cleaning 'respondent_name' column 
titles = ['Dr. ', 'Mr. ', 'Ms. ', 'Prof. ']
for title in titles:
    df['respondent_name'] = df['respondent_name'].str.replace(title, '', regex=False)
```

```python
print(df['respondent_name'].head(20))
```

```python
# fix capitalization
df['respondent_name'] = df['respondent_name'].str.title()
print(df['respondent_name'].head(20))
```

```python
print(df['feedback'].head(20))
```

```python
print(df['feedback'].value_counts(dropna=False))
```

```python
df['feedback'] = df['feedback'].str.strip()
```

```python
import numpy as np
df['feedback'] = df['feedback'].replace('No comment', np.nan)
```

```python
print(df['feedback'].isna().sum())
```

```python
# verify 
print("verify columns are clean")
print(f"\nGender unique: {df['gender'].unique()}")
print(f"\nEmployment status unique: {df['employment_status'].unique()}")
print(f"\nDepartment unique: {df['department'].unique()}")
print(f"\nSatisfaction unique: {df['satisfaction'].unique()}")
print(f"\nRecommend unique: {df['recommend'].unique()}")
print(f"\nMissing values:\n{df.isna().sum()}")
```

```python
df.to_csv('survey_data_clean.csv', index=False)
print("Cleaned data saved!")
```
