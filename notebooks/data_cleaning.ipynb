# Survey Response Data Cleaning

This notebook demonstrates systematic cleaning of text and categorical data.

## Methodology
For each categorical column: **Explore → Strip → Map → Verify**

```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('../data/raw/survey_data_messy.csv')
print(f"Shape: {df.shape}")
df.head()
```

## 1. Gender Column
Issues: Case inconsistency, abbreviations, whitespace

```python
# EXPLORE: Check unique values
print("Before cleaning:")
print(df['gender'].unique())
```

```python
# STRIP whitespace
df['gender'] = df['gender'].str.strip()

# MAP to standard values
gender_mapping = {
    'MALE': 'Male',
    'M': 'Male',
    'male': 'Male',
    'm': 'Male',
    'female': 'Female',
    'f': 'Female',
    'F': 'Female',
    'FEMALE': 'Female'
}
df['gender'] = df['gender'].replace(gender_mapping)

# VERIFY
print("After cleaning:")
print(df['gender'].unique())
```

## 2. Employment Status Column
Issues: Case inconsistency, format variations

```python
# EXPLORE
print("Before cleaning:")
print(df['employment_status'].unique())
```

```python
# STRIP + MAP
df['employment_status'] = df['employment_status'].str.strip()

employ_status_mapping = {
    'full-time': 'Full-time',
    'Full Time': 'Full-time',
    'Fulltime': 'Full-time',
    'full time': 'Full-time',
    'Part time': 'Part-time',
    'Contractor': 'Contract',
    'Freelance': 'Contract',
    'intern': 'Intern'
}
df['employment_status'] = df['employment_status'].replace(employ_status_mapping)

# VERIFY
print("After cleaning:")
print(df['employment_status'].unique())
```

## 3. Department Column (Most Complex)
Issues: Abbreviations, typos, suffixes, full names

```python
# EXPLORE
print("Before cleaning:")
print(df['department'].unique())
print(f"\n{df['department'].nunique()} unique values!")
```

```python
# STRIP + MAP
df['department'] = df['department'].str.strip()

department_mapping = {
    # Sales variations
    'sales': 'Sales',
    'Sales Team': 'Sales',
    'sales department': 'Sales',
    
    # Marketing variations
    'marketing': 'Marketing',
    'Mktg': 'Marketing',
    'Mkt': 'Marketing',
    'mktg': 'Marketing',
    'Mrkting': 'Marketing',
    'Marketng': 'Marketing',
    'marketing department': 'Marketing',
    
    # IT variations
    'I.T.': 'IT',
    'Tech': 'IT',
    'Information Technology': 'IT',
    'IT Department': 'IT',
    'information technology': 'IT',
    
    # HR variations
    'hr': 'HR',
    'Human Resources': 'HR',
    'Human Resource': 'HR',
    'HR Dept': 'HR',
    
    # R&D variations
    'R and D': 'R&D',
    'R & D': 'R&D',
    'Research & Development': 'R&D',
    'Research and Development': 'R&D',
    'Research': 'R&D'
}
df['department'] = df['department'].replace(department_mapping)

# VERIFY
print("After cleaning:")
print(df['department'].unique())
```

## 4. Satisfaction Column
Issues: Case inconsistency

```python
# STRIP + MAP
df['satisfaction'] = df['satisfaction'].str.strip()

satisfaction_mapping = {
    'very satisfied': 'Very Satisfied',
    'satisfied': 'Satisfied',
    'SATISFIED': 'Satisfied'
}
df['satisfaction'] = df['satisfaction'].replace(satisfaction_mapping)

# VERIFY
print(df['satisfaction'].unique())
```

## 5. Recommend Column
Issues: Case inconsistency, abbreviations

```python
# STRIP + MAP
df['recommend'] = df['recommend'].str.strip()

recommend_mapping = {
    'yes': 'Yes',
    'YES': 'Yes',
    'Y': 'Yes',
    'maybe': 'Maybe',
    'N': 'No',
    'no': 'No'
}
df['recommend'] = df['recommend'].replace(recommend_mapping)

# VERIFY
print(df['recommend'].unique())
```

## 6. Respondent Names
Issues: Titles (Dr., Mr., etc.), case inconsistency

```python
# Remove titles
titles = ['Dr. ', 'Mr. ', 'Ms. ', 'Mrs. ', 'Prof. ']
for title in titles:
    df['respondent_name'] = df['respondent_name'].str.replace(title, '', regex=False)

# Standardize to Title Case
df['respondent_name'] = df['respondent_name'].str.title()

# VERIFY
print(df['respondent_name'].head(10))
```

## 7. Feedback Column
Issues: Whitespace, meaningless values

```python
# STRIP whitespace
df['feedback'] = df['feedback'].str.strip()

# Convert "No comment" to NaN (not usable feedback)
df['feedback'] = df['feedback'].replace('No comment', np.nan)

# VERIFY
print(f"Missing feedback: {df['feedback'].isna().sum()}")
```

## Final Validation

```python
print("=== FINAL VALIDATION ===")
print(f"\nGender: {df['gender'].unique()}")
print(f"\nEmployment: {df['employment_status'].unique()}")
print(f"\nDepartment: {df['department'].unique()}")
print(f"\nSatisfaction: {df['satisfaction'].unique()}")
print(f"\nRecommend: {df['recommend'].unique()}")
print(f"\nMissing values:\n{df.isna().sum()}")
```

```python
# Save cleaned data
df.to_csv('../data/processed/survey_data_clean.csv', index=False)
print("Cleaned data saved!")
```
